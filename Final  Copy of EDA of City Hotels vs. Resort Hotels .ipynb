````python
import pandas as pd
import numpy as np

# Function to detect and calculate percentage of outliers using the IQR method
def calculate_outlier_percentage(df):
    outlier_percentage = {}

    # Loop through all numerical columns
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[column].quantile(0.25)  # First quartile (25%)
        Q3 = df[column].quantile(0.75)  # Third quartile (75%)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identifying outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        # Calculate percentage of outliers
        outlier_percentage[column] = (len(outliers) / len(df)) * 100

    return outlier_percentage

# Calculating the percentage of outliers for all numerical columns
outlier_percentage_all_columns = calculate_outlier_percentage(data)

# Display the outlier percentage for each numerical column
outlier_percentage_all_columns
````

````python
import pandas as pd
import numpy as np

# Function to remove outliers using the IQR method
def remove_outliers_iqr(df):
    # Loop through all numerical columns
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[column].quantile(0.25)  # First quartile (25%)
        Q3 = df[column].quantile(0.75)  # Third quartile (75%)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Removing outliers by keeping only values within the bounds
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df

# Remove outliers from the dataset
data_cleaned = remove_outliers_iqr(data)

# Display the number of rows after removing outliers
print(f"Number of rows after removing outliers: {len(data_cleaned)}")
````

#### Chart - 12

````python
# Chart - 12 visualization code
import pandas as pd

# Extract the month and year from the data
# Assuming there's already a 'arrival_date_month' and 'arrival_date_year' column
data['arrival_date'] = pd.to_datetime(data['arrival_date_year'].astype(str) + '-' + data['arrival_date_month'], format='%Y-%B')

# Group by month and count the number of bookings per month
monthly_bookings = data['arrival_date_month'].value_counts().sort_index()

# Plot the number of bookings per month
monthly_bookings.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Number of Bookings by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()

# To calculate seasons based on the month
def assign_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    elif month in ['September', 'October', 'November']:
        return 'Autumn'

# Add a 'season' column
data['season'] = data['arrival_date_month'].apply(assign_season)

# Group by season and count the number of bookings per season
seasonal_bookings = data['season'].value_counts()

# Plot the number of bookings per season
seasonal_bookings.plot(kind='bar', figsize=(8, 6), color='lightcoral')
plt.title('Number of Bookings by Season')
plt.xlabel('Season')
plt.ylabel('Number of Bookings')
plt.show()
````
